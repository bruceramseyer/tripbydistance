```python
#Data check for mixed-type values
for col in df_mixedtest.columns.tolist():
    weird=(df_mixedtest[[col]].applymap(type)!=df_mixedtest[[col]].iloc[0].apply(type)).any(axis=1)
    if len(df_mixedtest[weird])>0:
        print(col)
```

```python
#Convert df_test to string value
df_mixedtest['mix']=df_mixedtest['mix'].astype('str')
```

```python
#Removing null values from County impacted columns
df_trips_clean=df_trips.dropna(subset=['Population Staying at Home','Population Not Staying at Home',
                                       'Number of Trips','Number of Trips <1','Number of Trips 1-3',
                                      'Number of Trips 3-5','Number of Trips 5-10','Number of Trips 10-25',
                                      'Number of Trips 25-50','Number of Trips 50-100','Number of Trips 100-250',
                                      'Number of Trips 250-500','Number of Trips >=500'], how='all')
```
