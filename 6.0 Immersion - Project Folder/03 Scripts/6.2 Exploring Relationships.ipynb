<font size="4"> 2. Exploring Relations

```python
#Convert Level column into subgroups
df_nation=df_trips_clean[df_trips_clean['Level'] =='National']
df_state=df_trips_clean[df_trips_clean['Level'] =='State']
df_county=df_trips_clean[df_trips_clean['Level'] =='County']
```

```python
#Create subgroups for only numerical columns
subnum_nation=df_nation[['Population Not Staying at Home','Number of Trips <1','Number of Trips 1-3',
                         'Number of Trips 3-5','Number of Trips 5-10','Number of Trips 10-25','Number of Trips 25-50',
                         'Number of Trips 50-100','Number of Trips 100-250','Number of Trips 250-500','Number of Trips >=500']]
subnum_state=df_state[['Population Not Staying at Home','Number of Trips <1','Number of Trips 1-3',
                         'Number of Trips 3-5','Number of Trips 5-10','Number of Trips 10-25','Number of Trips 25-50',
                         'Number of Trips >=500']]
subnum_county=df_county[['Population Not Staying at Home','Number of Trips <1','Number of Trips 1-3',
                         'Number of Trips 3-5','Number of Trips 5-10','Number of Trips 10-25','Number of Trips 25-50',
                         'Number of Trips >=500']]
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr_nation = sns.heatmap(subnum_nation.corr(), annot = True, ax = ax)
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr_state = sns.heatmap(subnum_state.corr(), annot = True, ax = ax)
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr_county = sns.heatmap(subnum_county.corr(), annot = True, ax = ax)
```

<font size="3">The higher the variables are to 1.0 demonstrates a stronger correlation for Populations Not Staying at Home and the number of trips taken by distance. Based off the results at a National, State, and County level the strongest correlation on total trips by distance is 50 miles or less.

```python
# Create a scatterplot for randomized County "Population Not Staying at Home" and "Number of Trips >=500" columns in seaborn

sns.lmplot(x = 'Population Not Staying at Home', y = 'Number of Trips >=500', data = df_county_sm)
```

<font size="3"> The strongest correlation between populations not staying home and traveling between 3-5 miles was at the National level, both state and county stayed more under the trendline as the size of the population grew. 
    
<font size="3"> When comparing the population not staying home to the larger traveling distance of >=500 miles, the correlation was much weaker. Fewer number of trips occurred at all levels as the population size grew.

```python
# Creating a categorical plot of 'Population Not Staying at Home' at a National level for potential categorization

nothomenation_hist = sns.histplot(df_nation['Population Not Staying at Home'], bins = 20, kde = True)
```

```python
df_nation.loc[df_nation['Number of Trips'] < 1300000000, 'Trips category'] = 'Less trips'
```

```python
# Create a categorical plot in seaborn using the trip categories created above

sns.set(style="ticks")
g = sns.catplot(x="Number of Trips", y="Population Not Staying at Home", hue="Nat_Pop_Cat", data=df_nation)

plt.xscale('log')
```

```python
trip_threshold = df_nation[df_nation['Number of Trips'] < 1250000000]
```

```python
# Create a categorical plot in seaborn using the trip categories created above

sns.set(style="ticks")
g1 = sns.catplot(x="Number of Trips", y="Population Not Staying at Home", hue="Nat_Pop_Cat", data=trip_threshold)

plt.xscale('log')
```

<font size="3"> The categorical plot illustrates the obvious conclusion that the number of trips increases with the larger amount of population. However, we can see there is a threshold at the number of trips within the linear trend. Path that threshold, the population and number of trips flattens out. Limiting the Number of Trips to the rounded mean of 1,250,000,000 shows the upward trend peaks around that average
